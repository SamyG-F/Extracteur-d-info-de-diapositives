Building the corresponding json
---------------------

The objective is whatever the initial format, you identify the format, use the corresponding parser, then from the json data structure work.

Latex has got a hierarchical structure. Of course you can put the hierarchical information as features in a flat structure. 

```python
# https://docs.python.org/3.8/library/json.html
import json
from json import loads # from text to object
from json import dumps # from object to text

from pptx import Presentation

def parse_pptx_presentation (presentation):
    # Parsing a pptx presentation and return a json object
    # slides > paragraph > run
    #https://python-pptx.readthedocs.io/en/latest/api/presentation.html#pptx.presentation.Presentation
    prs = dict()

    if presentation.core_properties.title:
        if presentation.core_properties.title.strip() != '': prs['title'] = presentation.core_properties.title
    if presentation.core_properties.subject:
        if presentation.core_properties.subject.strip() != '': prs['subject'] = presentation.core_properties.subject
    if presentation.core_properties.keywords:
        if presentation.core_properties.keywords.strip() != '': prs['keywords'] = presentation.core_properties.keywords
    if presentation.core_properties.created:
        if presentation.core_properties.created != None: prs['created'] = str(presentation.core_properties.created)
    if presentation.core_properties.modified:
        if presentation.core_properties.modified != None: prs['modified'] = str(presentation.core_properties.modified)
    prs['length'] = len(presentation.slides)
    slides = []
    for i in range(0, len(presentation.slides)):
        #https://python-pptx.readthedocs.io/en/latest/api/slides.html#pptx.slide.Slide
        s = presentation.slides[i]
        slide = dict()
        # alternative: s.slide_id ; unique identifier whatever the position
        slide['page'] = str(i+1)
        if s.shapes.title != None:
            slide['header'] = str(s.shapes.title.text)
        else:
            slide['header'] = ''
        
        text_frames = []
        for shape in s.shapes:
            #print (shape)
            frame = dict()
            if shape.is_placeholder:
                #print('shape.is_placeholder')
                phf = shape.placeholder_format
                #print('shape.placeholder_format idx=%d, type=%s' % (phf.idx, phf.type))
                frame['type'] = str(phf.type)
                #print ("Presentation.slides[].placeholders[{}]={}".format(s.placeholders[phf.idx].text))
                #text_frame = shape.text_frame
            if shape.has_text_frame:
                    paragraphs = []
                    for paragraph in shape.text_frame.paragraphs:
                        runs = []
                        for run in paragraph.runs:
                            feats = dict ()
                            feats['text'] = run.text
                            if run.font.name != None: feats['font'] = run.font.name
                            if run.font.bold != None: feats['bold'] = run.font.bold
                            if run.font.italic != None: feats['italic'] = run.font.italic
                            if run.font.underline != None: feats['underline'] = run.font.underline
                            if run.font.size != None: feats['size'] = run.font.size    
                            if run.hyperlink.address != None: feats['href'] = run.hyperlink.address     #text_runs.append(run.text)
                            if run.font.color.type != None:
                                if run.font.color != None: 
                                    feats['color'] = str(run.font.color.rgb)
                            runs.append(feats)
                        paragraphs.append(runs)
                    frame['paragraphs'] = paragraphs
                    text_frames.append(frame)
                # Debug:
            #print ('Debug: frame=', frame)
        slide['frames'] = text_frames
        slides.append(slide)
        # Debug:
        #print ('Debug: slide=', slide)
    prs['slides'] = slides

    return prs

# json_obj to text
#json.dumps(prs, indent=2, sort_keys=True)

#print (json.dumps(parse_pptx_presentation(Presentation('data/CM1_React.pptx')), indent=2, sort_keys=True))
#print (json.dumps(parse_pptx_presentation(Presentation('data/CM3_MobxRouter_CM4_React_Avance.pptx')), indent=2, sort_keys=True))

```

## Normalize whitespace character 



```python
import re

def normalize_ws_character(string):
    # turn all the \s into ' '
    return re.sub('\s+', ' ', string)
```

```python
from pptx import Presentation
import json

# --------------------------------------------------------------------
def get_presentation_headers(presentation): 
    # Extracting slides headers from json presentation
    return [s['header'] for s in presentation['slides']]    

def get_paragraph_runs(paragraph):
    return [run['text'] for run in paragraph]

def get_paragraph_text(paragraph):
    return '\n'.join(get_paragraph_runs(paragraph))

def get_frame_runs(frame):
    texts = []
    for paragraph in frame['paragraphs']:
        texts.extend(get_paragraph_runs(paragraph))
    return texts

def get_frame_text(frame):
    return '\n'.join(get_frame_runs(frame))

def get_slide_runs(slide):
    texts = []
    for frame in slide['frames']:
        texts.extend(get_frame_runs(frame))
    return texts

def get_slide_text(slide):
    return '\n'.join(get_slide_runs(slide))

# --------------------------------------------------------------------
def get_presentation_runs(presentation):
    # get a list of the minimal textual units in the given presentation
    texts = []
    for slide in presentation['slides']:
        texts.extend(get_slide_runs(slide))
    return texts

# --------------------------------------------------------------------
def get_presentation_text(presentation):
    # get the presentation texts as one string
    return '\n'.join(get_presentation_runs(presentation))

# --------------------------------------------------------------------
def get_presentation_slides_text(presentation):
    # get a list of the slides text for the given presentation
    return [get_slide_text(slide) for slide in presentation['slides']]
    
# --------------------------------------------------------------------
def get_presentation_slide_bodies_text(presentation):
    # get a list of the slides text for the given presentation
    return [get_slide_text(slide) for slide in presentation['slides']]    

# --------------------------------------
#p_json = parse_pptx_presentation(Presentation('data/CM1_React.pptx'))
p_json = parse_pptx_presentation(Presentation('../exide/tests/data/pptx/presentation-test.pptx'))

print (json.dumps(p_json, indent=4, sort_keys=True))

# la liste des unités textuelles les plus élémentaires de la présentation
#print (get_presentation_runs(p_json))

# le texte dans sa mise en forme originale
#print (get_presentation_text(p_json))

# le texte en une seule ligne 
#print (normalize_ws_character(get_presentation_text(p_json)))
# --------------------------------------
# la liste des titres de la présentation
#print (get_presentation_headers(p_json))

# la liste des textes de chaque diapositive 
# print (get_presentation_slides_text(p_json))

# normalize ws of each presentation slides text
#print ([normalize_ws_character(s_t) for s_t in get_presentation_slides_text(p_json)])

# 
for slide in p_json['slides']:
    print ('Debug: get_slide_runs='+str(get_slide_runs(slide)))
    
# --------------------------------------
# normalize keywords
#from lib.nlp import normalize_sentence_to_words
#from lib.nlp import load_fr_stopwords

# init
#fr_stopwords = load_fr_stopwords()

#print ([normalize_sentence_to_words(fr_stopwords, normalize_ws_character(s_t)) for s_t in get_presentation_slides_text(p_json)])


```
